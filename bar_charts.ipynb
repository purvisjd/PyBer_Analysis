```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y-axis to a list of floats as the total fare in US dollars
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```

```python
#create the plot
plt.bar(x_axis, y_axis)
```

```python
#create the plot
plt.bar(x_axis, y_axis, color="green", label="Boston")

#create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#create title
plt.title("PyBer Fare by Month")

#add the legend
plt.legend()
```

```python
#create plot
plt.barh(x_axis, y_axis)
```

```python
#create plot with data on opposite axes
plt.barh(y_axis, x_axis)
```

```python
#invert to have months in ascending order
#gca() means "get current axes"
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#skill drill
#create plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()

#add legend
plt.legend()

#add title
plt.title("PyBer Fare by Month")

#add axes labels
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
#create vertical bar chart with object oriented method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#create horizontal bar chart with object oriented method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
#skills test
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.legend()
```
