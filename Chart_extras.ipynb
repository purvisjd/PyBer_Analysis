```python
%matplotlib inline

#import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
#set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y-axis to a list of floats as the fares
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```

```python
#get the standard deviation of the values in the y_axis

stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
#adding caps to error bars
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#adding error bars using objec oriented approach
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#adding error bars to bar chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#adding error bars to bar chart with object oriented approach
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#modifying ticks 
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
#modifying ticks using object oriented approach
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
#set minor ticks to x-axis
#to use set_minor_locator function and multiplelocator method, must import multiplelocator from matplotlib.ticker
from matplotlib.ticker import MultipleLocator

#increase size of plot figure
fig, ax = plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

#create minor ticks at increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
