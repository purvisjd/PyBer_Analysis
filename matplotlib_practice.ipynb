```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 35.42, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label="Boston")

#create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit between 0 and 45
plt.ylim(0, 45)

#create a title
plt.title("PyBer Fare by Month")

#add the legend
plt.legend()
```

```python
#create the plot
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")

#create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit between 0 and 45
plt.ylim(0,45)

#create title
plt.title("PyBer Fare by Month")

#add a grdi
plt.grid()

#add the legend
plt.legend()
```
