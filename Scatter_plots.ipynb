```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y-axis to a list of floats as the total fare in US dollars
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

```

```python
#create scatter plot 
plt.plot(x_axis, y_axis, 'o')
```

```python
#create scatter plot
plt.scatter(x_axis, y_axis)
```

```python
#skill test
plt.scatter(y_axis, x_axis, color="red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0,50)
plt.legend()
plt.gca().invert_yaxis()
```

```python
#creating a bubble chart
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#change sizes of markers in graph
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
#use new data for s parameter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#using list comprehension to replace loop
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#create scatterplot using object oriented interface

fix, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#create bubble chart using object oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
#skill drill; use object oriented approach to make changes
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color="skyblue", label="Boston", alpha=0.2, linewidths=2, edgecolors="black", s = [i * 5 for i in y_axis])
ax.set_title("PyBer Fare By Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.legend()
ax.invert_yaxis()
plt.xlim(0,50)

```
